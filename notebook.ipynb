```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('x02.csv', names=['gender', 'age', 'head_size', 'brain_weight'])
dataset.head()
```

```python
x = dataset['brain_weight'].values * 1.0
y = dataset['head_size'].values + 1.0
print(x.dtype)
print(y.dtype)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title('Simple plot')
ax.set_xlabel('Brain weight')
ax.set_ylabel('Head size')
```

```python
polynomCoeff = np.polyfit(x, y, deg=1)
polynomCoeff2 = np.polyfit(x, y, deg=2)
print(polynomCoeff)
```

```python
h = np.poly1d(polynomCoeff)
h2 = np.poly1d(polynomCoeff2)
print(h)
print(h2)
```

```python
h(1200)
```

```python
x_h = np.linspace(np.amin(x), np.amax(x), 1000)
x_h2 = np.linspace(np.amin(x), np.amax(x), 1000)
print(x_h)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x_h, h(x_h), color='r')
ax.plot(x_h2, h2(x_h2), color='g')
ax.set_title('Simple plot')
ax.set_xlabel('Brain weight')
ax.set_ylabel('Head size')
```
